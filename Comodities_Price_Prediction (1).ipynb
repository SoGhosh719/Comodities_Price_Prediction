Autocorrelation and Partial Autocorrelation

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Ensure the index is datetime
df.index = pd.to_datetime(df.index)

# Set a classy Seaborn theme
sns.set_theme(style="whitegrid")

# Define custom function for ACF and PACF plots
def plot_acf_pacf(ts_data, commodity_name):
    fig, axes = plt.subplots(1, 2, figsize=(14, 6), dpi=120)

    # Autocorrelation plot
    plot_acf(ts_data, lags=30, ax=axes[0], zero=False, color='teal', alpha=0.8)
    axes[0].set_title(f'{commodity_name} - Autocorrelation (ACF)', fontsize=14)
    axes[0].set_xlabel('Lags', fontsize=12)
    axes[0].set_ylabel('ACF Value', fontsize=12)

    # Partial Autocorrelation plot
    plot_pacf(ts_data, lags=30, ax=axes[1], zero=False, color='darkorange', alpha=0.8)
    axes[1].set_title(f'{commodity_name} - Partial Autocorrelation (PACF)', fontsize=14)
    axes[1].set_xlabel('Lags', fontsize=12)
    axes[1].set_ylabel('PACF Value', fontsize=12)

    # Improve spacing
    plt.tight_layout()

    # Show the plots
    plt.show()

# Loop through each commodity and generate the plots
for commodity in df['commodity'].unique():
    ts_data = df[df['commodity'] == commodity]['close'].dropna()
    plot_acf_pacf(ts_data, commodity)
```

## Performing Time Series Analysis

**Performing Arima Diagnostics**

```python
from statsmodels.tsa.arima.model import ARIMA

# Define a function for better visual diagnostics
def plot_arima_diagnostics(results, commodity):
    residuals = results.resid
    fig, axes = plt.subplots(2, 2, figsize=(12, 10), constrained_layout=True)

    # Residuals plot
    sns.lineplot(x=residuals.index, y=residuals, ax=axes[0, 0], color="steelblue")
    axes[0, 0].set_title(f'{commodity} - Residuals Over Time', fontsize=14)
    axes[0, 0].set_xlabel('Date')
    axes[0, 0].set_ylabel('Residuals')

    # Density plot of residuals
    sns.histplot(residuals, kde=True, color="teal", ax=axes[0, 1])
    axes[0, 1].set_title(f'{commodity} - Residuals Density Plot', fontsize=14)
    axes[0, 1].set_xlabel('Residuals')
    axes[0, 1].set_ylabel('Density')

    # Q-Q plot
    from statsmodels.graphics.gofplots import qqplot
    qqplot(residuals, line="s", ax=axes[1, 0], color="purple", alpha=0.6)
    axes[1, 0].set_title(f'{commodity} - Q-Q Plot', fontsize=14)

    # ACF of residuals
    from statsmodels.graphics.tsaplots import plot_acf
    plot_acf(residuals, lags=30, ax=axes[1, 1], color="orangered", alpha=0.7)
    axes[1, 1].set_title(f'{commodity} - ACF of Residuals', fontsize=14)

    # General styling
    for ax in axes.flatten():
        ax.grid(True, linestyle='--', alpha=0.6)

    plt.suptitle(f'{commodity} - ARIMA Diagnostics', fontsize=16, weight="bold")
    plt.show()
```

```python
# Iterate through each commodity
for commodity in df['commodity'].unique():
    ts_data = df[df['commodity'] == commodity]['close'].dropna()
    model = ARIMA(ts_data, order=(1, 1, 1))
    results = model.fit()

    print(f'{commodity} - ARIMA Summary:')
    print(results.summary())

    # Plot improved diagnostics
    plot_arima_diagnostics(results, commodity)
```

**Evaluation of the Performance of the models**

```python
import warnings
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Suppress warnings to keep the output clean
warnings.filterwarnings("ignore")

for commodity in df['commodity'].unique():
    ts_data = df[df['commodity'] == commodity]['close'].dropna()
    train_size = int(len(ts_data) * 0.8)
    train, test = ts_data[:train_size], ts_data[train_size:]

    # Fit ARIMA model
    model = ARIMA(train, order=(1, 1, 1))
    results = model.fit()

    # Forecast and evaluate
    predictions = results.forecast(steps=len(test))
    rmse = np.sqrt(mean_squared_error(test, predictions))
    mae = mean_absolute_error(test, predictions)

    print(f'{commodity} - RMSE: {rmse}, MAE: {mae}')
```

```python
from prophet import Prophet

# Set seaborn style for better visual aesthetics
sns.set_theme(style="whitegrid")

# Loop through each unique commodity in the dataset
for commodity in df['commodity'].unique():
    # Prepare the data for Prophet
    prophet_data = df[df['commodity'] == commodity].reset_index()[['date', 'close']]
    prophet_data.rename(columns={'date': 'ds', 'close': 'y'}, inplace=True)

    # Create the Prophet model
    model = Prophet()
    model.fit(prophet_data)

    # Create future dataframe for prediction
    future = model.make_future_dataframe(periods=365)
    forecast = model.predict(future)

    # Create a figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    # Plot the actual price on the first subplot
    ax1.plot(prophet_data['ds'], prophet_data['y'], color='blue', label='Actual Price')
    ax1.set_title(f'{commodity} - Actual Prices', fontsize=14, weight='bold', color='darkblue')
    ax1.set_xlabel('Date', fontsize=12, weight='bold')
    ax1.set_ylabel('Price', fontsize=12, weight='bold')
    ax1.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
    ax1.legend()

    # Plot the predicted price (forecast) on the second subplot
    ax2.plot(forecast['ds'], forecast['yhat'], color='green', label='Predicted Price')
    ax2.fill_between(forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'], color='orange', alpha=0.3, label='Uncertainty Interval')
    ax2.set_title(f'{commodity} - Predicted Prices with Prophet', fontsize=14, weight='bold', color='darkgreen')
    ax2.set_xlabel('Date', fontsize=12, weight='bold')
    ax2.set_ylabel('Price', fontsize=12, weight='bold')
    ax2.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
    ax2.legend()

    # Adjust layout to prevent overlapping
    plt.tight_layout()

    # Show the plot
    plt.show()
```

### Using LSTM

```python
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error
```

```python
merged_df['date'] = pd.to_datetime(merged_df['date'])  # Ensure 'date' is in datetime format
merged_df = merged_df.set_index('date')  # Set 'date' as the index
```

```python
merged_df['ticker'].unique()
```

```python
# Get all unique tickers
unique_commodity = merged_df['commodity'].unique()
```

```python
# Loop over each ticker and perform LSTM training and prediction
for idx, commodity in enumerate(unique_commodity):

    print("For",commodity)

    # Filter the data for the current ticker
    merged_df_commodity = merged_df[merged_df['commodity'] == commodity]

    # Use 'close' as the target
    prices = merged_df_commodity['close'].values

    # Normalize the data (MinMax scaling)
    scaler = MinMaxScaler(feature_range=(0, 1))
    prices_scaled = scaler.fit_transform(prices.reshape(-1, 1))

    # Function to create the dataset with look-back for time series prediction
    def create_dataset(data, look_back=60):
        X, y = [], []
        for i in range(look_back, len(data)):
            X.append(data[i-look_back:i, 0])  # Features for the LSTM
            y.append(data[i, 0])  # Target price
        return np.array(X), np.array(y)

    look_back = 60  # Use previous 60 days for prediction
    X, y = create_dataset(prices_scaled, look_back)

    # Reshape X to be 3D as required by LSTM: [samples, time steps, features]
    X = X.reshape(X.shape[0], X.shape[1], 1)

    # Step 2: Split the data into train and test sets
    train_size = int(len(X) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    # Step 3: Build the LSTM model
    model = Sequential([
            LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
            LSTM(units=50, return_sequences=False),
            Dropout(0.2),
            Dense(units=25),
            Dense(units=1)  # Predict the next closing price
        ])

    model.compile(optimizer='adam', loss='mean_squared_error')

    # Step 4: Train the model
    model.fit(X_train, y_train, batch_size=32, epochs=20)

    # Step 5: Predict using the trained LSTM model
    predictions_scaled = model.predict(X_test)

    # Inverse scaling to get the actual values
    predictions = scaler.inverse_transform(predictions_scaled)
    actual_prices_y = scaler.inverse_transform(y_test.reshape(-1, 1))

    # Calculate the accuracy metrics
    mape = mean_absolute_percentage_error(actual_prices_y, predictions)
    rmse = np.sqrt(mean_squared_error(actual_prices_y, predictions))

    print(f"Accuracy Metrics for {commodity}:")
    print(f"MAPE: {mape * 100:.2f}%")
    print(f"RMSE: {rmse:.2f}")

    # Convert to pandas DataFrame for better handling
    predicted_prices = pd.DataFrame({'date': merged_df_commodity.reset_index()['date'].iloc[-len(predictions_scaled):], 'predicted': predictions.flatten()})
    actual_prices = pd.DataFrame({'date': merged_df_commodity.reset_index()['date'].iloc[-len(predictions_scaled):], 'actual': actual_prices_y.flatten()})

    # Set Seaborn style for a classy look
    sns.set_theme(style="whitegrid")

    # Create a figure and axes
    fig, ax = plt.subplots(1, 2, figsize=(14, 6))

    # Plot Actual Prices
    sns.lineplot(data=actual_prices, x='date', y='actual', ax=ax[0], color='blue', label='Actual Price')
    ax[0].set_title(commodity)
    ax[0].set_xlabel("Date")
    ax[0].set_ylabel("Price")
    ax[0].tick_params(axis='x', rotation=45)

    # Plot Predicted Prices
    sns.lineplot(data=predicted_prices, x='date', y='predicted', ax=ax[1], color='red', label='Predicted Price')
    ax[1].set_title(commodity)
    ax[1].set_xlabel("Date")
    ax[1].set_ylabel("Price")
    ax[1].tick_params(axis='x', rotation=45)

    # Adjust layout and show the plot
    plt.tight_layout()
    plt.show()
```
